# WeatherPy

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

from api_keys import api_key

from citipy import citipy
```

```python
output_data_file = "output/cities_data.csv"
```

```python
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)
```

```python
print("We have " + str(len(cities)) + " cities.")
```

```python
city_df = pd.DataFrame({"City" : cities})

city_df["Country"] = ""
city_df["Cloudiness"] = ""
city_df["Humidity"] = ""
city_df["Max Temp"] = ""
city_df["Wind Speed"] = ""
city_df["Lat"] = ""
city_df["Lng"] = ""

#city_df.head()
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}appid={api_key}&units={units}&q="

#query_url
```

```python
start_time = time.time()
request_number = 1

for index, row in city_df.iterrows():
    
    city_name = row["City"]
    response = requests.get(query_url + city_name).json()
    print(f"Retrieving results for city #{index + 1} of {len(city_df)}: {city_name}")
    
    try:      
        city_df.loc[index, "Country"] = response["sys"]["country"]
        city_df.loc[index, "Cloudiness"] = response["clouds"]["all"]
        city_df.loc[index, "Humidity"] = response["main"]["humidity"]
        city_df.loc[index, "Max Temp"] = response["main"]["temp_max"]
        city_df.loc[index, "Wind Speed"] = response["wind"]["speed"]
        city_df.loc[index, "Lat"] = response["coord"]["lat"]
        city_df.loc[index, "Lng"] = response["coord"]["lon"]
    
    except KeyError:
        print("ERROR: City not found/ information missing. Skipping...")  
    
    print("------------")
    
    request_number = request_number + 1
    if request_number == 55:
        current_time = time.time()
        sleep_time = current_time - start_time
        time.sleep(60 - sleep_time)
        start_time = time.time()
        request_number = 1
```

```python
#city_df
```

```python
null_rows = city_df.loc[city_df["Country"] == ""]
#len(null_rows)
city_df_final = city_df.drop(city_df.index[null_rows.index.values])
city_df_final = city_df_final.reset_index(drop=True)
print("We now have " + str(len(city_df_final)) + " cities.")
```

## Three observable trends based on the data:

* The relationship between humidity and latitude is weak, however a very large number of cities can be seen to have 100% humidity. From this collection of cities, higher humidity is more common than lower humidity (clear majority of the cities have over 60% humidity, and very few are below 40% humidity).
* The relationship between cloudiness and latitude is also weak. It is important to note that there are very large groupings of cities that have 0%, 75%, or around 90% cloudiness. It appears that the majority of cities fall into one of these categories, with 0% cloudiness having by far the most cities. There are some other weak groupings of cities at 20% and 40% cloudiness.
* Max temperature peaks when the latitude is roughly between 0 and 20. This is logical because overall temperatures are warmer closer to the equator. Tying in with this, max temperatures decrease as the latitude increases and decreases (moves away from the equator). This also makes sense since overall temperatures are colder closer to the poles. Additionally, negative latitudes seem to be warmer than positive latitudes, where the lowest max temperature the negative latitudes reach is around 40 degrees Fahrenheit and the lowest max temperature the positive latitudes reach is primarily around 30 degrees Fahrenheit.
